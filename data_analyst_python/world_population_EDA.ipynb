### renaming columns

```python
df_pd = pd.DataFrame(df)
df_pd.rename(columns = {'2022 Population':'2022_population'}, inplace = True)
df_pd.rename(columns = {'2020 Population':'2020_population'}, inplace = True)
df_pd.rename(columns = {'2015 Population':'2015_population'}, inplace = True)
df_pd.rename(columns = {'2010 Population':'2010_population'}, inplace = True)
df_pd.rename(columns = {'2000 Population':'2000_population'}, inplace = True)
df_pd.rename(columns = {'1990 Population':'1990_population'}, inplace = True)
df_pd.rename(columns = {'1980 Population':'1980_population'}, inplace = True)
df_pd.rename(columns = {'1970 Population':'1970_population'}, inplace = True)
df_pd.rename(columns = {'Area (km²)':'area'}, inplace = True)
df_pd.rename(columns = {'Density (per km²)':'density'}, inplace = True)
df_pd.rename(columns = {'Growth Rate':'growth_rate'}, inplace = True)
df_pd.rename(columns = {'World Population Percentage':'world_population_percentage'}, inplace = True)
```

```python
df_pd.columns
```

### Top 10 most populated places.

```python
df_pd.sort_values(by='world_population_percentage', ascending=False).head(10)
```

```python
top_10 = df_pd.sort_values(by='world_population_percentage', ascending=False).head(10)

plt.figure(figsize=(17, 6))
sns.barplot(x='Country', y='world_population_percentage', data=top_10)
plt.title("Top 10 most populated places in the world")
plt.show()
```

### Population counts grouped by Continents

```python
region_totals = df_pd.groupby('Continent')['2022_population'].agg(['sum', 'count'])
region_totals
```

```python
plt.figure(figsize=(17, 6))
sns.barplot(x=region_totals.index, y='sum', data=region_totals)
plt.show()
```
